```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

## • use the data from the file, read it to a dataframe and make any modification that you may find useful to solve the task

```python
df = pd.read_csv('../data/ab-test.csv')
df
```

```python
res1 = df[df['time'] == 'before']
data_plot_1 = res1[res1['group'] == 'control']['diff']
data_plot_2 = res1[res1['group'] == 'test']['diff']
data_plot_before = [data_plot_1, data_plot_2]
res2 = df[df['time'] == 'after']
data_plot_3 = res2[res2['group'] == 'control']['diff']
data_plot_4 = res2[res2['group'] == 'test']['diff']
data_plot_after = [data_plot_3, data_plot_4]
```

## • the figsize is still the same, you can choose whatever fontsize you like
## • the color palette should be the same as in the example
## • the fontsize of the title is 15
## • the width of the box lines is 3, the width of the median lines is 2

```python
style_data =  {'boxprops': dict(color='g', linewidth=3),
               'capprops': dict(color='r'),
               'whiskerprops': dict(color='r'),
               'medianprops': dict(color='r', linewidth=2)
               }


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 8), sharey='all')
plt.rc('font', size=13)
axes[0].boxplot(data_plot_after, showmeans=True, labels=['control', 'test'], **style_data)
axes[0].grid(axis='both')
axes[1].grid(axis='both')
axes[1].boxplot(data_plot_before, showmeans=True, labels=['control', 'test'], **style_data)
axes[0].set(title='before')
axes[0].set_xlabel('[group]')
axes[1].set_xlabel('[group]')
axes[1].set(title='after')
fig.suptitle('Lag before labs deadlines. After and before Newsfeed', fontsize=15)
```
